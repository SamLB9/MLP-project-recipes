```python
# PAS TOUCHER
# MAIN GET DATA TO JSON
import json
import numpy as np
import pandas as pd
import re
from recipe_scrapers import scrape_me

def get_alims(items):
    # Expression régulière mise à jour pour capturer les nouvelles unités
    pattern = re.compile(r'(\d+\.?\d*\s*(?:mg|g|kg|cl|l|ml|c\. à café|pincée\(s\)|graine\(s\)|c\. à soupe|Gousse\(s\)|Gousse|gousse\(s\)|graine\(s\)|brin\(s\)|branche\(s\)|boule|portion\(s\)|pincée\(s\)|tranche\(s\)|poignée\(s\)|botte\(s\)|verre\(s\)|boîte\(s\)|pot\(s\)|bouquet\(s\)|sachet\(s\)|feuilles|gousse|Bocal \/ bocaux)?)\s+(.*)')
    output_dict = {}

    for item in items:
        item = re.sub(r'Gousse\(s\)|Gousse|gousse\(s\)|gousse', 'gousse(s)', item)
        # Gérer les cas où un ingrédient est une combinaison (comme "Sel & Poivre")
        if '&' in item:
            combined_ingredients = item.split('&')
            for ingredient in combined_ingredients:
                ingredient = ingredient.strip().capitalize()
                # Supprime les prépositions comme "de" du nom de l'ingrédient
                ingredient = re.sub(r'^de\s+', '', ingredient)
                # Si l'ingrédient n'est pas déjà dans le dictionnaire ou si la quantité actuelle est moins spécifique, mettez à jour
                if ingredient not in output_dict or output_dict[ingredient] == '1 u':
                    output_dict[ingredient] = '1 u'
            continue

        match = pattern.match(item)
        if match:  # Si un modèle avec une quantité est trouvé
            quantity = match.group(1).strip()  # Enlève les espaces inutiles
            ingredient = match.group(2).strip()  # Enlève les espaces inutiles

            # Supprime les prépositions comme "de" du nom de l'ingrédient
            ingredient = re.sub(r'^de\s+', '', ingredient)

            # Met en majuscule le premier mot de l'ingrédient
            ingredient = ingredient.capitalize()

            # Attribue '1 u' si l'unité est l'une de celles spécifiées
            if any(unit in quantity for unit in ["pincée(s)", "graine(s)", "boule", "portion(s)", "poignée(s)", "verre(s)"]):
                output_dict[ingredient] = '1 u'
            else:
                output_dict[ingredient] = quantity
        else:  # Pour les éléments sans quantité spécifique
            # Met en majuscule le premier mot de l'élément
            item = item.capitalize()
            output_dict[item] = '1 u'

    return output_dict


def adjust_values(dico, real, future):
    pattern = re.compile(r'(\d+(\.\d+)?)')  # Modifié pour inclure les nombres décimaux
    
    for key, value in dico.items():
        # Si la valeur est '1 u', la transformer en '1'
        if value == '1 u':
            # dico[key] = '1'
            continue

        str_value = str(value)
        matches = pattern.findall(str_value)
        if matches:
            number_str = matches[0][0]
            number = float(number_str) if '.' in number_str else int(number_str)
            
            calculated_value = (number * future) / real
            
            calculated_value = round(calculated_value, 3)  # Arrondi à 2 décimales

            new_value_str = str_value.replace(number_str, str(calculated_value), 1)
            dico[key] = new_value_str

    return dico


def get_data(plat, pers, alims):
    data = {
        'Plat':plat,
        'Nbre de pers': pers,
        'Alims': alims
    }
    return data


# Retourne toute la data d'une recette pour tous nbres de personne(s) 
def get_data_adjust(plat, pers, alims):
    alims = get_alims(alims)
    all_data = []

    # Plage de personnes pour lesquelles ajuster la recette
    person_range = range(1, 9) if pers > 4 else range(1, pers + 5)

    for p in person_range:
        adjusted_alims = adjust_values(dict(alims), pers, p)
        all_data.append(get_data(plat, p, adjusted_alims))

    return all_data


def test_scrap(list, file):
    file_path = file
    with open(file_path, 'w') as fichier_json:
        fichier_json.write("")
    all_recipes = []

    for url in list:
        if url is not None:
            scraper = scrape_me(url)
            try:
                plat = scraper.title()
                alims = scraper.ingredients()
                pers = scraper.yields()
                pers.split(' ')
                pers = pers[0]
                pers = int(pers)
                if pers > 2:
                    data = get_data_adjust(plat, pers, alims)
                    all_recipes.extend(data)
                    with open(file_path, 'a') as json_file:
                        json.dump(data, json_file, indent=4)
            except TypeError:
                pass

    with open(file_path, 'w') as json_file:
        json.dump(all_recipes, json_file, indent=4)
```

```python
def create_dataframes_from_json(json_data):
    # Créer une liste vide pour stocker les DataFrames
    dataframes_list = []

    # Parcourir chaque valeur possible de "Nbre de pers" (1 à 8)
    for nbre_de_pers in range(1, 9):
        # Filtrer les éléments du JSON pour ceux avec la valeur spécifique de "Nbre de pers"
        filtered_data = [item for item in json_data if item["Nbre de pers"] == nbre_de_pers]

        if filtered_data:
            # Collecter tous les aliments uniques à partir de tous les plats
            all_aliments = set()
            for item in filtered_data:
                all_aliments.update(item["Alims"].keys())

            # Convertir en liste pour maintenir l'ordre
            aliments = list(all_aliments)

            # Créer un DataFrame à partir des données filtrées
            df = pd.DataFrame(filtered_data)

            # Initialiser un dictionnaire pour les nouvelles colonnes
            new_columns = {aliment: [] for aliment in aliments}

            # Remplir le dictionnaire avec les données d'aliments
            for aliment in aliments:
                new_columns[aliment] = df["Alims"].apply(lambda x: x.get(aliment, 0))

            # Créer un nouveau DataFrame à partir du dictionnaire des nouvelles colonnes
            aliments_df = pd.DataFrame(new_columns)

            # Concaténer avec le DataFrame original (sans la colonne "Alims")
            df = pd.concat([df.drop('Alims', axis=1), aliments_df], axis=1)

            # Ajouter ce DataFrame à la liste
            dataframes_list.append(df)

    return dataframes_list





file_path = 'recipes_v2.json'

# Ouvrir le fichier JSON et charger les données
with open(file_path, 'r') as file:
    json_data = json.load(file)

dataframes_list = create_dataframes_from_json(json_data)
```

```python
final_df = pd.read_csv('Data/Old_Data/recipes_1.csv', sep=';')

def consolidate_columns(df):
    # Get base column names (without numeric suffixes)
    base_col_names = set(col.rsplit('.', 1)[0] for col in df.columns if '.' in col)
    # Create a new DataFrame to store the consolidated columns
    consolidated_df = pd.DataFrame(index=df.index)
    
    # Iterate over each base column name
    for base_name in base_col_names:
        # Get a list of all columns for this base name
        col_list = [col for col in df.columns if col.startswith(base_name)]
        # Start with the first column as the base
        consolidated_col = df[col_list[0]].astype(str)
        # Consolidate with the subsequent columns
        for col in col_list[1:]:
            consolidated_col = consolidated_col.where(consolidated_col != '0', df[col].astype(str))
        
        # Add the consolidated column to the new DataFrame
        consolidated_df[base_name] = consolidated_col
    
    # Drop the old columns from the original DataFrame
    df = df.drop(columns=[col for col in df.columns if '.' in col])
    # Join the original DataFrame with the new consolidated DataFrame
    df = df.join(consolidated_df, rsuffix='_consolidated')
    
    # If original base name column exists, it may be duplicated, so we remove it
    for base_name in base_col_names:
        if base_name in df.columns:
            df.drop(base_name, axis=1, inplace=True)
            # Rename the consolidated column to the original base name
            df.rename(columns={base_name + '_consolidated': base_name}, inplace=True)
    
    return df

# Apply the function to the DataFrame
final_df = consolidate_columns(final_df)

# Display the first few rows of the final DataFrame
final_df['Ail'].head(40)
```

```python
# Conversion csv source en 10 csv pour differents nombre de personne
import pandas as pd
import re
import math

def convertir_unites(quantite, unite):
    # Conversion systématique des unités g, ml et cl en kg ou L
    conversions = {'g': 1000, 'ml': 1000, 'cl': 100}
    nouvelle_unite = {'g': 'kg', 'ml': 'l', 'cl': 'l'}
    
    if unite in conversions:
        return quantite / conversions[unite], nouvelle_unite[unite]
    return quantite, unite


def ajuster_cellule(cellule, multiplicateur):
    cellule_str = str(cellule).strip()
    if cellule_str == '0':
        return '0'

    # Utiliser une expression régulière pour extraire les quantités et les unités
    match = re.match(r"(\d+\.?\d*)\s*(g|ml|cl|kg|l|pod\(s\)?|slice\(s\)?|bunch\(s\)?|leaf\(s\)?|box\(s\)?|sprig\(s\)?)", cellule_str, re.IGNORECASE)
    
    if match:
        quantite, unite = match.groups()
        quantite_modifiee = float(quantite) * multiplicateur

        # Appeler la fonction de conversion
        quantite_modifiee, unite_modifiee = convertir_unites(quantite_modifiee, unite.lower())

        # Arrondir la quantité sauf pour 'kg' et 'l'
        if unite_modifiee not in ['kg', 'l']:
            quantite_modifiee = math.ceil(quantite_modifiee) if not quantite_modifiee.is_integer() else int(quantite_modifiee)

        if unite_modifiee in ['kg', 'l']:
            quantite_modifiee = round(quantite_modifiee, 3)
        
        # Traitement des unités avec "(s)"
        if "(s" in unite:
            unite_modifiee = unite_modifiee.replace("(s", "(s)")

        return f"{quantite_modifiee} {unite_modifiee}"
    else:
        # Traitement des cellules qui ne correspondent pas au format quantité + unité
        match_nombre_seul = re.match(r"^\d+\.?\d*$", cellule_str)
        if match_nombre_seul:
            quantite_modifiee = float(cellule_str) * multiplicateur
            return str(math.ceil(quantite_modifiee))
        else:
            return cellule_str



def ajuster_recettes(df):
    liste_df = []
    for nb_personnes in range(1, 11):
        df_modifie = df.copy()
        df_modifie['No. people'] = nb_personnes
        for col in df.columns:
            if col != 'No. people':
                df_modifie[col] = df[col].apply(lambda x: ajuster_cellule(x, nb_personnes)).astype(df[col].dtype)
        liste_df.append(df_modifie)
    return liste_df


final_df = pd.read_csv('Data/test1_eng.csv', sep=';')

def consolidate_columns(df):
    # Get base column names (without numeric suffixes)
    base_col_names = set(col.rsplit('.', 1)[0] for col in df.columns if '.' in col)
    # Create a new DataFrame to store the consolidated columns
    consolidated_df = pd.DataFrame(index=df.index)
    
    # Iterate over each base column name
    for base_name in base_col_names:
        # Get a list of all columns for this base name
        col_list = [col for col in df.columns if col.startswith(base_name)]
        # Start with the first column as the base
        consolidated_col = df[col_list[0]].astype(str)
        # Consolidate with the subsequent columns
        for col in col_list[1:]:
            consolidated_col = consolidated_col.where(consolidated_col != '0', df[col].astype(str))
        
        # Add the consolidated column to the new DataFrame
        consolidated_df[base_name] = consolidated_col
    
    # Drop the old columns from the original DataFrame
    df = df.drop(columns=[col for col in df.columns if '.' in col])
    # Join the original DataFrame with the new consolidated DataFrame
    df = df.join(consolidated_df, rsuffix='_consolidated')
    
    # If original base name column exists, it may be duplicated, so we remove it
    for base_name in base_col_names:
        if base_name in df.columns:
            df.drop(base_name, axis=1, inplace=True)
            # Rename the consolidated column to the original base name
            df.rename(columns={base_name + '_consolidated': base_name}, inplace=True)
    
    return df

# Apply the function to the DataFrame
final_df = consolidate_columns(final_df)

scale_dfs = ajuster_recettes(final_df)

scale_sorted_dfs = []

for df in scale_dfs:
    first_two_columns = df.iloc[:, :2]

    # Sort the rest of the columns alphabetically
    sorted_columns = sorted(df.columns[2:])
    sorted_df = df[sorted_columns]

    # Concatenate the first two columns with the sorted ones
    df = pd.concat([first_two_columns, sorted_df], axis=1)
    scale_sorted_dfs.append(df)

name = ['Data/test1_eng.csv', 'Data/test2_eng.csv', 'Data/test3_eng.csv', 'Data/test4_eng.csv', 'Data/test5_eng.csv', 'Data/test6_eng.csv', 'Data/test7_eng.csv', 'Data/test8_eng.csv', 'Data/test9_eng.csv', 'Data/test10_eng.csv',]

for i in range(len(scale_sorted_dfs)):
    scale_sorted_dfs[i].to_csv(name[i], index=False, sep=';')

```
